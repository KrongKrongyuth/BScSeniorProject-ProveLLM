```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
vanila_path = {'all-minilm-theorem':None, 'all-minilm-default':None,
               'bge-m3-theorem':None,     'bge-m3-default':None}
finetune_path = {'all-minilm-theorem':None, 'all-minilm-default':None}
```

```python
vanila_path['all-minilm-theorem'] = './all-minilm-l6-v2_w-theorem/all-minilm-l6-v2_w-theorem-retrieve-perf.csv'
vanila_path['all-minilm-default'] = './all-minilm-l6-v2_wo-theorem/all-minilm-l6-v2_wo-theorem-retrieve-perf.csv'
finetune_path['all-minilm-theorem'] = './all-minilm-I6-v2-finetune_w-theorem/all-minilm-I6-v2-finetune_w-theorem-retrieve-perf.csv'
finetune_path['all-minilm-default'] = './all-minilm-I6-v2-finetune_wo-theorem/all-minilm-I6-v2-finetune_wo-theorem-retrieve-perf.csv'
```

```python
fine_th = pd.read_csv(finetune_path['all-minilm-theorem']).drop('Unnamed: 0', axis=1)
fine_df = pd.read_csv(finetune_path['all-minilm-default']).drop('Unnamed: 0', axis=1)
mini_th = pd.read_csv(vanila_path['all-minilm-theorem']).drop('Unnamed: 0', axis=1)
mini_df = pd.read_csv(vanila_path['all-minilm-default']).drop('Unnamed: 0', axis=1)
```

```python
# Create subplots
fig, axes = plt.subplots(4, 4, figsize=(15, 10))

# Plot histograms for P@3 from the four datasets
sns.histplot(fine_df['P@3'], ax=axes[0, 0], color='blue')
axes[0, 0].set_title('fine-tune model without theorem')

sns.histplot(fine_th['P@3'], ax=axes[1, 0], color='green')
axes[1, 0].set_title('fine-tune model with theorem')

sns.histplot(mini_df['P@3'], ax=axes[2, 0], color='red')
axes[2, 0].set_title('All-minilm-L6-V2 without theorem')

sns.histplot(mini_th['P@3'], ax=axes[3, 0], color='orange')
axes[3, 0].set_title('All-minilm-L6-V2 with theorem')

# Plot histograms for P@5 from the four datasets
sns.histplot(fine_df['P@5'], ax=axes[0, 1], color='blue')
axes[0, 1].set_title('fine-tune model without theorem')

sns.histplot(fine_th['P@5'], ax=axes[1, 1], color='green')
axes[1, 1].set_title('fine-tune model with theorem')

sns.histplot(mini_df['P@5'], ax=axes[2, 1], color='red')
axes[2, 1].set_title('All-minilm-L6-V2 without theorem')

sns.histplot(mini_th['P@5'], ax=axes[3, 1], color='orange')
axes[3, 1].set_title('All-minilm-L6-V2 with theorem')

# Plot histograms for P@10 from the four datasets
sns.histplot(fine_df['P@10'], ax=axes[0, 2], color='blue')
axes[0, 2].set_title('fine-tune model without theorem')

sns.histplot(fine_th['P@10'], ax=axes[1, 2], color='green')
axes[1, 2].set_title('fine-tune model with theorem')

sns.histplot(mini_df['P@10'], ax=axes[2, 2], color='red')
axes[2, 2].set_title('All-minilm-L6-V2 without theorem')

sns.histplot(mini_th['P@10'], ax=axes[3, 2], color='orange')
axes[3, 2].set_title('All-minilm-L6-V2 with theorem')

# Plot histograms for P@50 from the four datasets
sns.histplot(fine_df['P@50'], ax=axes[0, 3], color='blue')
axes[0, 3].set_title('fine-tune model without theorem')

sns.histplot(fine_th['P@50'], ax=axes[1, 3], color='green')
axes[1, 3].set_title('fine-tune model with theorem')

sns.histplot(mini_df['P@50'], ax=axes[2, 3], color='red')
axes[2, 3].set_title('All-minilm-L6-V2 without theorem')

sns.histplot(mini_th['P@50'], ax=axes[3, 3], color='orange')
axes[3, 3].set_title('All-minilm-L6-V2 with theorem')

# Set common labels
fig.suptitle('Histogram of Precision for Four Datasets')
for idx, ax in enumerate(axes.flat):
    if idx % 4 == 0:
        ax.set_xlabel(f'P@3')
    elif idx % 4 == 1:
        ax.set_xlabel(f'P@5')
    elif idx % 4 == 2:
        ax.set_xlabel(f'P@10')
    elif idx % 4 == 3:
        ax.set_xlabel(f'P@50')
        
    ax.set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
# Create subplots
fig, axes = plt.subplots(4, 4, figsize=(15, 10))

# Plot histograms for R@3 from the four datasets
sns.histplot(fine_df['R@3'], ax=axes[0, 0], color='blue')
axes[0, 0].set_title('fine-tune model without theorem')

sns.histplot(fine_th['R@3'], ax=axes[1, 0], color='green')
axes[1, 0].set_title('fine-tune model with theorem')

sns.histplot(mini_df['R@3'], ax=axes[2, 0], color='red')
axes[2, 0].set_title('All-minilm-L6-V2 without theorem')

sns.histplot(mini_th['R@3'], ax=axes[3, 0], color='orange')
axes[3, 0].set_title('All-minilm-L6-V2 with theorem')

# Plot histograms for R@5 from the four datasets
sns.histplot(fine_df['R@5'], ax=axes[0, 1], color='blue')
axes[0, 1].set_title('fine-tune model without theorem')

sns.histplot(fine_th['R@5'], ax=axes[1, 1], color='green')
axes[1, 1].set_title('fine-tune model with theorem')

sns.histplot(mini_df['R@5'], ax=axes[2, 1], color='red')
axes[2, 1].set_title('All-minilm-L6-V2 without theorem')

sns.histplot(mini_th['R@5'], ax=axes[3, 1], color='orange')
axes[3, 1].set_title('All-minilm-L6-V2 with theorem')

# Plot histograms for R@10 from the four datasets
sns.histplot(fine_df['R@10'], ax=axes[0, 2], color='blue')
axes[0, 2].set_title('fine-tune model without theorem')

sns.histplot(fine_th['R@10'], ax=axes[1, 2], color='green')
axes[1, 2].set_title('fine-tune model with theorem')

sns.histplot(mini_df['R@10'], ax=axes[2, 2], color='red')
axes[2, 2].set_title('All-minilm-L6-V2 without theorem')

sns.histplot(mini_th['R@10'], ax=axes[3, 2], color='orange')
axes[3, 2].set_title('All-minilm-L6-V2 with theorem')

# Plot histograms for R@50 from the four datasets
sns.histplot(fine_df['R@50'], ax=axes[0, 3], color='blue')
axes[0, 3].set_title('fine-tune model without theorem')

sns.histplot(fine_th['R@50'], ax=axes[1, 3], color='green')
axes[1, 3].set_title('fine-tune model with theorem')

sns.histplot(mini_df['R@50'], ax=axes[2, 3], color='red')
axes[2, 3].set_title('All-minilm-L6-V2 without theorem')

sns.histplot(mini_th['R@50'], ax=axes[3, 3], color='orange')
axes[3, 3].set_title('All-minilm-L6-V2 with theorem')

# Set common labels
fig.suptitle('Histogram of Recall for Four Datasets')
for idx, ax in enumerate(axes.flat):
    if idx % 4 == 0:
        ax.set_xlabel(f'R@3')
    elif idx % 4 == 1:
        ax.set_xlabel(f'R@5')
    elif idx % 4 == 2:
        ax.set_xlabel(f'R@10')
    elif idx % 4 == 3:
        ax.set_xlabel(f'R@50')
        
    ax.set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
# Create a new dataframe to hold all the precision values
data_precision = {
    'Model': ['fine-tune without theorem'] * len(fine_df) + ['fine-tune with theorem'] * len(fine_th) +
             ['All-minilm-L6-V2 without theorem'] * len(mini_df) + ['All-minilm-L6-V2 with theorem'] * len(mini_th),
    'P@3': pd.concat([fine_df['P@3'], fine_th['P@3'], mini_df['P@3'], mini_th['P@3']]),
    'P@5': pd.concat([fine_df['P@5'], fine_th['P@5'], mini_df['P@5'], mini_th['P@5']]),
    'P@10': pd.concat([fine_df['P@10'], fine_th['P@10'], mini_df['P@10'], mini_th['P@10']]),
    'P@50': pd.concat([fine_df['P@50'], fine_th['P@50'], mini_df['P@50'], mini_th['P@50']])
}

df_precision = pd.DataFrame(data_precision)

# Melt the dataframe for easier plotting
df_precision_melted = df_precision.melt(id_vars='Model', var_name='Metric', value_name='Value')

# Create the precision boxplot
plt.figure(figsize=(15, 5))
sns.boxplot(x='Metric', y='Value', hue='Model', data=df_precision_melted)
plt.title('Comparison of Precision between Datasets')
plt.xticks(rotation=45)
plt.show()

# Create a new dataframe to hold all the recall values
data_recall = {
    'Model': ['fine-tune without theorem'] * len(fine_df) + ['fine-tune with theorem'] * len(fine_th) +
             ['All-minilm-L6-V2 without theorem'] * len(mini_df) + ['All-minilm-L6-V2 with theorem'] * len(mini_th),
    'R@3': pd.concat([fine_df['R@3'], fine_th['R@3'], mini_df['R@3'], mini_th['R@3']]),
    'R@5': pd.concat([fine_df['R@5'], fine_th['R@5'], mini_df['R@5'], mini_th['R@5']]),
    'R@10': pd.concat([fine_df['R@10'], fine_th['R@10'], mini_df['R@10'], mini_th['R@10']]),
    'R@50': pd.concat([fine_df['R@50'], fine_th['R@50'], mini_df['R@50'], mini_th['R@50']])
}

df_recall = pd.DataFrame(data_recall)

# Melt the dataframe for easier plotting
df_recall_melted = df_recall.melt(id_vars='Model', var_name='Metric', value_name='Value')

# Create the recall boxplot
plt.figure(figsize=(15, 5))
sns.boxplot(x='Metric', y='Value', hue='Model', data=df_recall_melted)
plt.title('Comparison of Recall between Datasets')
plt.xticks(rotation=45)
plt.show()
```

```python
# Calculate the number of 0s in RR for each dataset
zero_counts_rr = {
    'Model': ['fine-tune without theorem', 'fine-tune with theorem', 'All-minilm-L6-V2 without theorem', 'All-minilm-L6-V2 with theorem'],
    'RR': [fine_df['RR'].eq(0).sum(), fine_th['RR'].eq(0).sum(), mini_df['RR'].eq(0).sum(), mini_th['RR'].eq(0).sum()]
}

df_zero_counts_rr = pd.DataFrame(zero_counts_rr)

# Create the bar chart for RR
plt.figure(figsize=(10, 6))
barplot_rr = sns.barplot(x='Model', y='RR', data=df_zero_counts_rr, hue='Model')
plt.title('Comparison of Zero Counts in RR between Datasets')
plt.xlabel('Model')
plt.ylabel('Zero Counts')
plt.xticks(rotation=45)

# Display the count on the top of the bar
for p in barplot_rr.patches:
    barplot_rr.annotate(format(p.get_height(), '.0f'), 
                        (p.get_x() + p.get_width() / 2., p.get_height()), 
                        ha = 'center', va = 'center', 
                        xytext = (0, 9), 
                        textcoords = 'offset points')

plt.show()
```

```python
df_join = pd.merge(
    left=mini_df,
    right=fine_df,
    how='left',
    on='theorem_id')

zero_zero = df_join[df_join['RR_x'].eq(0) & df_join['RR_y'].eq(0)][['theorem_id', 'theorem_contents_x', 'RR_x', 'RR_y']]
zero_value = df_join[df_join['RR_x'].eq(0) & ~df_join['RR_y'].eq(0)][['theorem_id', 'theorem_contents_x', 'RR_x', 'RR_y']]
value_zero = df_join[~df_join['RR_x'].eq(0) & df_join['RR_y'].eq(0)][['theorem_id', 'theorem_contents_x', 'RR_x', 'RR_y']]

compare_rr = {'Model': ['Zero then Zero', 'Zero then Non-zero', 'None-zero then zero'],
              'RR': [len(zero_zero), len(zero_value), len(value_zero)]}
compare_rr = pd.DataFrame(compare_rr)

plt.figure(figsize=(10, 6))
barplot_rr = sns.barplot(x='Model', y='RR', data=compare_rr, hue='Model')
plt.title('Comparison of before and after in RR (without theorem dataset)\nVanilla vs Fine-tune')
plt.xlabel('Model')
plt.ylabel('Counts')
plt.xticks(rotation=45)

# Display the count on the top of the bar
for p in barplot_rr.patches:
    barplot_rr.annotate(format(p.get_height(), '.0f'), 
                        (p.get_x() + p.get_width() / 2., p.get_height()), 
                        ha = 'center', va = 'center', 
                        xytext = (0, 9), 
                        textcoords = 'offset points')

plt.show()
```
